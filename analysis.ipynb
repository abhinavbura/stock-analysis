**Plot the daily, weekly and monthly returns of a set of stocks of your choice** 
* references:
    * https://www.investopedia.com/terms/y/yearly-rate-of-return-method.asp
    * https://www.fool.com/knowledge-center/how-to-calculate-a-monthly-return-on-investment/#:~:text=Take%20the%20ending%20balance%2C%20and,corresponds%20to%20your%20monthly%20return.

```python
for company in companies_id:
    company['Daily Return'] = company['Adj Close']/(company['Adj Close'].shift(1))-1

# Then we'll plot the daily return percentage
fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(15)
fig.set_figwidth(20)

AAPL['Daily Return'].plot(ax=axes[0,0], legend=True)
axes[0,0].set_title('APPLE')

GOOG['Daily Return'].plot(ax=axes[0,1], legend=True)
axes[0,1].set_title('GOOGLE')

MSFT['Daily Return'].plot(ax=axes[1,0], legend=True)
axes[1,0].set_title('MICROSOFT')

AMZN['Daily Return'].plot(ax=axes[1,1], legend=True)
axes[1,1].set_title('AMAZON')

TSLA['Daily Return'].plot(ax=axes[2,0], legend=True)
axes[2,0].set_title('TESLA')

fig.delaxes(axes[2][1])
fig.tight_layout()
```

```python
for company in companies_id:
    company['Weekly Return'] = company['Adj Close']/(company['Adj Close'].shift(7))-1
# Then we'll plot the daily return percentage
fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(15)
fig.set_figwidth(20)

AAPL['Weekly Return'].plot(ax=axes[0,0], legend=True)
axes[0,0].set_title('APPLE')

GOOG['Weekly Return'].plot(ax=axes[0,1], legend=True)
axes[0,1].set_title('GOOGLE')

MSFT['Weekly Return'].plot(ax=axes[1,0], legend=True)
axes[1,0].set_title('MICROSOFT')

AMZN['Weekly Return'].plot(ax=axes[1,1], legend=True)
axes[1,1].set_title('AMAZON')

TSLA['Weekly Return'].plot(ax=axes[2,0], legend=True)
axes[2,0].set_title('TESLA')

fig.delaxes(axes[2][1])
fig.tight_layout()
```

```python
for company in companies_id:
    company['Monthly Return'] = company['Adj Close']/(company['Adj Close'].shift(30))-1
# Then we'll plot the daily return percentage
fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(15)
fig.set_figwidth(20)

AAPL['Monthly Return'].plot(ax=axes[0,0], legend=True)
axes[0,0].set_title('APPLE')

GOOG['Monthly Return'].plot(ax=axes[0,1], legend=True)
axes[0,1].set_title('GOOGLE')

MSFT['Monthly Return'].plot(ax=axes[1,0], legend=True)
axes[1,0].set_title('MICROSOFT')

AMZN['Monthly Return'].plot(ax=axes[1,1], legend=True)
axes[1,1].set_title('AMpython plot table for 3d matixAZON')

TSLA['Monthly Return'].plot(ax=axes[2,0], legend=True)
axes[2,0].set_title('TESLA')

fig.delaxes(axes[2][1])
fig.tight_layout()
```

Linear Regression is a statistical method that is used to predict the value of an output variable based on one or more input variables. In the context of stock market prediction, linear regression can be used to model the relationship between the historical prices of a stock and other financial indicators such as earnings, dividends, and market index.

LSTM (Long Short-Term Memory) is a type of Recurrent Neural Network (RNN) that is specifically designed to handle sequential data. LSTM can be used to predict stock prices by analyzing the historical prices of a stock and other financial indicators such as earnings, dividends, and market index.

Random Forest is an ensemble learning method that combines multiple decision trees to improve the accuracy of the predictions. Random Forest can be used to predict stock prices by analyzing the historical prices of a stock and other financial indicators such as earnings, dividends, and market index. The algorithm will build multiple decision trees and take the average of their predictions.

In summary, Linear Regression, LSTM and Random Forest are all machine learning algorithms that can be used to predict stock prices. They all work by analyzing the historical prices of a stock and other financial indicators, but each algorithm has its own strengths and weaknesses. Linear Regression is simple and easy to implement, but its predictions may not be accurate for complex data. LSTM and Random Forest are more powerful algorithms that can handle complex data and make accurate predictions, but they require more computational resources and are more difficult to implement.
